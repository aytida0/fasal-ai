```python
import kagglehub
import os
import shutil

# 1. Let kagglehub download the dataset to its lightning-fast hidden cache
print("Downloading New Plant Diseases Dataset...")
cache_path = kagglehub.dataset_download("vipoooool/new-plant-diseases-dataset")
print(f"Data initially cached at: {cache_path}")

# 2. Define your exact target directory and the crops you want
target_dir = '/content/Project_Datasets/PlantVillage'
target_crops = ['Corn', 'Tomato', 'Potato']
os.makedirs(target_dir, exist_ok=True)

# 3. Scan the cache and copy ONLY the required folders
print(f"Extracting targeted crop directories to {target_dir}...")
for root, dirs, files in os.walk(cache_path):
    for folder_name in dirs:
        # Check if the folder name contains 'Corn', 'Tomato', or 'Potato'
        if any(crop in folder_name for crop in target_crops):
            source_folder = os.path.join(root, folder_name)
            dest_folder = os.path.join(target_dir, folder_name)

            # Copy the folder to the target directory if it hasn't been copied yet
            if not os.path.exists(dest_folder):
                shutil.copytree(source_folder, dest_folder)
                print(f"Successfully staged: {folder_name}")

print(f"\nSuccess! Corn, Tomato, and Potato data is fully staged in: {target_dir}")
```

```python
import os

staging_dir = '/content/Project_Datasets'
fixed_count = 0

print("Scanning for corrupted filenames with trailing spaces...")
for root, dirs, files in os.walk(staging_dir):
    for filename in files:
        # Check if there is a space right before the file extension
        name, ext = os.path.splitext(filename)
        if name.endswith(' '):
            old_filepath = os.path.join(root, filename)

            # Remove the trailing space from the name
            new_name = name.rstrip() + ext
            new_filepath = os.path.join(root, new_name)

            os.rename(old_filepath, new_filepath)
            fixed_count += 1

print(f"Sanitation complete! Fixed {fixed_count} corrupted filenames.")
```

```python
import os
import shutil
from pathlib import Path

MASTER = "/content/MasterDataset"

# ---------- CLASS MAPPING ----------

mapping = {

    # CORN
    "Corn_(maize)___Common_rust_": "Corn___Common_Rust",
    "Corn_(maize)___Cercospora_leaf_spot Gray_leaf_spot": "Corn___Gray_Leaf_Spot",
    "Corn_(maize)___Northern_Leaf_Blight": "Corn___Northern_Leaf_Blight",
    "Corn_(maize)___healthy": "Corn___Healthy",

    # POTATO
    "Potato___Early_blight": "Potato___Early_Blight",
    "Potato___Late_blight": "Potato___Late_Blight",
    "Potato___healthy": "Potato___Healthy",

    # TOMATO
    "Tomato___Bacterial_spot": "Tomato___Bacterial_Spot",
    "Tomato___Early_blight": "Tomato___Early_Blight",
    "Tomato___Late_blight": "Tomato___Late_Blight",
    "Tomato___Leaf_Mold": "Tomato___Leaf_Mold",
    "Tomato___Septoria_leaf_spot": "Tomato___Septoria_Leaf_Spot",
    "Tomato___Tomato_mosaic_virus": "Tomato___Tomato_Mosaic_Virus",
    "Tomato___Tomato_Yellow_Leaf_Curl_Virus": "Tomato___Yellow_Leaf_Curl_Virus",
    "Tomato___healthy": "Tomato___Healthy",

    # RICE
    "Rice___Brown_Spot": "Rice___Brown_Spot",
    "Rice___Leaf_Blast": "Rice___Leaf_Blast",
    "Rice___Neck_Blast": "Rice___Neck_Blast",
    "Rice___Healthy": "Rice___Healthy",

    # WHEAT
    "Wheat___Brown_Rust": "Wheat___Brown_Rust",
    "Wheat___Yellow_Rust": "Wheat___Yellow_Rust",
    "Wheat___Healthy": "Wheat___Healthy",

    # SUGARCANE
    "Mosaic": "Sugarcane___Mosaic",
    "RedRot": "Sugarcane___Red_Rot",
    "Rust": "Sugarcane___Rust",
    "Yellow": "Sugarcane___Yellow",
    "Healthy": "Sugarcane___Healthy",

    # COTTON
    "bacterial_blight": "Cotton___Bacterial_Blight",
    "curl_virus": "Cotton___Curl_Virus",
    "fussarium_wilt": "Cotton___Fusarium_Wilt",
    "healthy": "Cotton___Healthy",
}

# ---------- DATASET ROOTS ----------

roots = [
    "/content/Project_Datasets/PlantVillage",
    "/content/Project_Datasets/FiveCrop/Rice",
    "/content/Project_Datasets/FiveCrop/Wheat",
    "/content/Project_Datasets/Sugarcane",
    "/content/Project_Datasets/Cotton/cotton"
]

# ---------- MERGE ----------

copied = 0

for root in roots:

    for folder in os.listdir(root):

        src_folder = os.path.join(root, folder)

        if not os.path.isdir(src_folder):
            continue

        if folder not in mapping:
            continue

        target_class = mapping[folder]
        dst_folder = os.path.join(MASTER, target_class)

        os.makedirs(dst_folder, exist_ok=True)

        for file in os.listdir(src_folder):

            if not file.lower().endswith(
                (".jpg", ".jpeg", ".png")
            ):
                continue

            src = os.path.join(src_folder, file)

            # unique filename
            new_name = f"{Path(root).name}_{file}"

            dst = os.path.join(dst_folder, new_name)

            shutil.copy2(src, dst)

            copied += 1

print(f"\nTotal Images Copied: {copied}")
```

```python
import os
import shutil
import random

SOURCE = "/content/MasterDataset"
DEST = "/content/dataset"

TRAIN = os.path.join(DEST, "train")
VAL   = os.path.join(DEST, "val")
TEST  = os.path.join(DEST, "test")

random.seed(42)

for split in [TRAIN, VAL, TEST]:
    os.makedirs(split, exist_ok=True)

for class_name in os.listdir(SOURCE):

    class_path = os.path.join(SOURCE, class_name)

    if not os.path.isdir(class_path):
        continue

    images = [
        f for f in os.listdir(class_path)
        if f.lower().endswith(
            (".jpg", ".jpeg", ".png")
        )
    ]

    random.shuffle(images)

    total = len(images)

    train_count = int(total * 0.75)
    val_count   = int(total * 0.10)

    train_imgs = images[:train_count]
    val_imgs   = images[train_count:train_count+val_count]
    test_imgs  = images[train_count+val_count:]

    for split_name, split_imgs in [
        ("train", train_imgs),
        ("val", val_imgs),
        ("test", test_imgs)
    ]:

        dst_dir = os.path.join(
            DEST,
            split_name,
            class_name
        )

        os.makedirs(dst_dir, exist_ok=True)

        for img in split_imgs:

            src = os.path.join(class_path, img)
            dst = os.path.join(dst_dir, img)

            shutil.copy2(src, dst)

print("Dataset Split Complete")
```

```python
import tensorflow as tf

IMG_SIZE = (224, 224)
BATCH_SIZE = 32

train_ds = tf.keras.utils.image_dataset_from_directory(
    "/content/dataset/train",
    image_size=IMG_SIZE,
    batch_size=BATCH_SIZE,
    label_mode="int",
    shuffle=True,
    seed=42
)

val_ds = tf.keras.utils.image_dataset_from_directory(
    "/content/dataset/val",
    image_size=IMG_SIZE,
    batch_size=BATCH_SIZE,
    label_mode="int",
    shuffle=False
)

test_ds = tf.keras.utils.image_dataset_from_directory(
    "/content/dataset/test",
    image_size=IMG_SIZE,
    batch_size=BATCH_SIZE,
    label_mode="int",
    shuffle=False
)
```

```python
class_names = train_ds.class_names

print("Number of classes:", len(class_names))
print()

for i, cls in enumerate(class_names):
    print(i, cls)
```

```python
AUTOTUNE = tf.data.AUTOTUNE

train_ds = train_ds.prefetch(AUTOTUNE)
val_ds = val_ds.prefetch(AUTOTUNE)
test_ds = test_ds.prefetch(AUTOTUNE)
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(12, 12))

for images, labels in train_ds.take(1):

    for i in range(9):

        ax = plt.subplot(3, 3, i + 1)

        plt.imshow(images[i].numpy().astype("uint8"))

        plt.title(class_names[labels[i]])

        plt.axis("off")

plt.show()
```

```python
data_augmentation = tf.keras.Sequential([
    tf.keras.layers.RandomFlip("horizontal"),
    tf.keras.layers.RandomRotation(0.15),
    tf.keras.layers.RandomZoom(0.15),
    tf.keras.layers.RandomContrast(0.10),
    tf.keras.layers.RandomTranslation(
        height_factor=0.10,
        width_factor=0.10
    )
])
```

```python
base_model = tf.keras.applications.MobileNetV3Small(
    input_shape=(224, 224, 3),
    include_top=False,
    weights="imagenet"
)

base_model.trainable = False
```

```python
inputs = tf.keras.Input(shape=(224, 224, 3))

x = data_augmentation(inputs)

x = tf.keras.applications.mobilenet_v3.preprocess_input(x)

x = base_model(x, training=False)

x = tf.keras.layers.GlobalAveragePooling2D()(x)

x = tf.keras.layers.Dropout(0.3)(x)

outputs = tf.keras.layers.Dense(
    33,
    activation="softmax"
)(x)

model = tf.keras.Model(inputs, outputs)
```

```python
model.compile(
    optimizer=tf.keras.optimizers.Adam(
        learning_rate=1e-4
    ),
    loss="sparse_categorical_crossentropy",
    metrics=["accuracy"]
)
```

```python
from sklearn.utils.class_weight import compute_class_weight
import numpy as np
import os

train_dir = "/content/dataset/train"

labels = []

for idx, cls in enumerate(class_names):

    cls_path = os.path.join(train_dir, cls)

    count = len([
        f for f in os.listdir(cls_path)
        if f.lower().endswith(
            (".jpg", ".jpeg", ".png")
        )
    ])

    labels.extend([idx] * count)

class_weights = compute_class_weight(
    class_weight="balanced",
    classes=np.unique(labels),
    y=labels
)

class_weights = {
    i: float(w)
    for i, w in enumerate(class_weights)
}

print(class_weights)
```

```python
callbacks = [

    tf.keras.callbacks.EarlyStopping(
        monitor="val_accuracy",
        patience=5,
        restore_best_weights=True
    ),

    tf.keras.callbacks.ModelCheckpoint(
        "best_model.keras",
        monitor="val_accuracy",
        save_best_only=True
    ),

    tf.keras.callbacks.ReduceLROnPlateau(
        monitor="val_loss",
        factor=0.2,
        patience=2,
        verbose=1
    )
]
```

```python
history = model.fit(
    train_ds,
    validation_data=val_ds,
    epochs=25,
    class_weight=class_weights,
    callbacks=callbacks
)
```
